```
import pandas as pd
import io
import requests
url="https://raw.githubusercontent.com/lnsc/datasetrandom/master/data.csv"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```
import numpy as np
from os import path
from PIL import Image
import os
os.getcwd()
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```
text = c.a[1]
#type(text[0])
#text[1]+" "+text[2]
text
```

```
# Creando y generando una imagen de nube de palabras.
wordcloud = WordCloud().generate(text)

wordcloud = WordCloud(max_font_size=50, max_words=200, background_color="Green").generate(text)
plt.figure()

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
#Guardando la imagen
wordcloud.to_file("WordCloud1.png")
```

```
# Creamos un array de una imagen de mascara
wine_mask = np.array(Image.open("33702.png"))
wine_mask

def transform_format(val):
    if val == 0:
        return 0
    else:
        return 255
```

```
transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]), np.int32)

for i in range(len(wine_mask)):
  transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
  
transformed_wine_mask
```

```
wc = WordCloud(background_color="black", max_words=300, mask=transformed_wine_mask,
              contour_width=3, contour_color='white')
#font_path="C:\Windows\Fonts\gulim.ttc"
# Generate a wordcloud
wc.generate(text)

image_colors = ImageColorGenerator(wine_mask)
# store to file
#wc.to_file("img2.png")

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
